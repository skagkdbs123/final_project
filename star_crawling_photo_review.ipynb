```python
brand='yale'
start = 1
url_list = []

page=1
while (page<3): #1페이지~2페이지
        try:
            url = f'https://www.musinsa.com/brands/{brand}?category3DepthCodes=&category2DepthCodes=&category1DepthCode=&colorCodes=&startPrice=&endPrice=&exclusiveYn=&includeSoldOut=&saleGoods=&timeSale=&includeKeywords=&sortCode=REVIEW&tags=&page={page}&size=90&listViewType=small&campaignCode=&groupSale=&outletGoods=&boutiqueGoods='            
            response = requests.get(url)
            soup = bs(response.text, 'lxml')

            for soup in soup.find_all('a', attrs={'class':'img-block'}):
                url_list.append(soup['href'].split('/')[-1])
            
            page += 1

        except:
            print(page)
            break

num = url_list
num
```

```python
title_list=[]
review_list=[]
rating_list=[]
body_list=[]
category_list=[]
num_list=[]
level_list=[]
name_list=[]
size_list=[]

page=1
i=0   #몇번째 상품부터 시작할건지

while (i<100):  # n-1 번째 상품까지할건지
    page=1  # 페이지 초기화
    while (page < 31):
        ###### goods_est_asc =낮은별점순 /// goods_est_desc 높은별점순 !!!!주의!!!!
        url = f'https://goods.musinsa.com/api/goods/v2/review/photo/list?similarNo=1727788&sort=goods_est_asc&selectedSimilarNo=0&page={page}&goodsNo={num[i]}&_=1671776931446'
        req = Request(url, headers={'User-Agent': 'Chrome/105.0.0.0 '})
        html = urlopen(req).read()
        soup = bs(html, 'html.parser')
        
        title= soup.select('a.review-goods-information__name')
        review= soup.select('div.review-contents__text')
        rating= soup.select('.review-list__rating__active')
        body= soup.select('div.review-profile__information')
        name= soup.select('p.review-profile__name')
        size= soup.select('span.review-goods-information__option')
        
        for t in title:
            title_list.append(t.text.strip())
        for b in body:
            body_list.append(b.select('p.review-profile__body_information'))
        for x in review:
            num_list.append(num[i])
        for r in review:
            review_list.append(r.text.strip())
        for p in rating:
            rating_list.append(p.attrs['style'].split()[1])
        for l1 in name:
            try:
                level_list.append(l1.text.strip().split()[1])
            except Exception:
                pass
        for l2 in name:
            try:
                name_list.append(l2.text.strip().split()[2])
            except Exception:
                pass
        for s in size:
            size_list.append(s.text.strip())
            
        
        #카테고리
        url2=f'https://www.musinsa.com/app/goods/{num[i]}'
        req2 = Request(url2, headers={'User-Agent': 'Chrome/105.0.0.0 '})
        html2 = urlopen(req2).read()
        soup2 = bs(html2, 'html.parser')
        cate = soup2.select_one('p.item_categories').get_text().replace('\n', '').replace(' ', '')
        
        for y in review:
            category_list.append(cate)
            
        df = pd.DataFrame([x for x in zip(title_list, review_list, rating_list, body_list, size_list, name_list, level_list, category_list, num_list)])
        df.columns = ['상품', '리뷰', '평점', '신체', 'size', '닉네임', '레벨', '카테고리', '상품_num']
        df.to_csv(f'{brand}_{i}번째 상품-asc-photo.csv', encoding='utf-8-sig', index=False)
        print(num[i], page)
        page += 1
    i +=1
```
